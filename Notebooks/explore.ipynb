```python
from torchcraft import replayer
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
folder_name = "../replay_data/1/"
file_list = os.listdir(folder_name)
file_list = [f for f in file_list if "TvT" in f]
file_list = [folder_name+f for f in file_list]
```

```python
## Explore all properties, attributes on one file first ##
replay = file_list[0]
print("Loading replay file..")
rep = replayer.load(replay)
```

```python
## Explore a frame ##
numFrames = len(rep)
numTeams = len(rep.getFrame(0).units.keys())
numUnits = np.zeros((numFrames, numTeams))
numResources = np.zeros((numFrames, numTeams, 4))

## To see teams ##
print("Teams: ", rep.getFrame(0).units.keys()); print("Team 0, Team 1, map")

for i in range(numFrames):
    frame = rep.getFrame(i)
    #print("On frame " + str(i))

    ### UNITS ###
    for k, (team,units) in enumerate(frame.units.iteritems()):
        numUnits[i,k] = len(units)

        if k == 0 and i == 0: ## just print once
            print("Frame " + str(i))
            ## Properties of units ##
            ## unit orders
            print("units->orders->first_frame: ", units[k].orders[k].first_frame)
            print("units->orders->type", units[k].orders[k].type)
            print("units->orders->targetId: ", units[k].orders[k].targetId)
            print("units->orders->targetX: ", units[k].orders[k].targetX)
            print("units->orders->targetY: ", units[k].orders[k].targetY)

            ## unit command
            print("units->command->frame: ", units[k].command.frame)
            print("units->command->type: ", units[k].command.type)
            print("units->command->targetId: ", units[k].command.targetId)
            print("units->command->targetX: ", units[k].command.targetX)
            print("units->command->targetY: ", units[k].command.targetY)
            print("units->command->extra:", units[k].command.extra)

            ## Other properties ##
            print("units->id: ", units[k].id) ## id
            print("units->x: ", units[k].x)
            print("units->y: ", units[k].y)
            print("units->flags: ", units[k].flags)
            print("units->health ", units[k].health)
            print("units->max_health ", units[k].max_health)
            print("units->shield ", units[k].shield)
            print("units->max_shield ", units[k].max_shield)
            print("units->energy ", units[k].energy)
            print("units->maxCD ", units[k].maxCD)
            print("units->groundCD ", units[k].groundCD)
            print("units->airCD ", units[k].airCD)
            print("units->visible ", units[k].visible)
            print("units->type: ", units[k].type)
            print("units->armor: ", units[k].armor)
            print("units->shieldArmor: ", units[k].shieldArmor)
            print("units->size: ", units[k].size)
            print("units->pixel_x: ", units[k].pixel_x)
            print("units->pixel_y: ", units[k].pixel_y)
            print("units->pixel_size_x: ", units[k].pixel_size_x)
            print("units->pixel_size_y: ", units[k].pixel_size_y)
            print("units->groundATK: ", units[k].groundATK)
            print("units->airATK: ", units[k].airATK)
            print("units->groundDmgType: ", units[k].groundDmgType)
            print("units->airDmgType: ", units[k].airDmgType)
            print("units->groundRange: ", units[k].groundRange)
            print("units->airRange: ", units[k].airRange)
            #print("units->orders: ", units[k].orders)
            #print("units->command: ", units[k].command)
            print("units->velocityX: ", units[k].velocityX)
            print("units->velocityY: ", units[k].velocityY)
            print("units->playerId: ", units[k].playerId)
            print("units->resources: ", units[k].resources)

            print("units->accelerating: ", units[k].accelerating)
            print("units->attacking: ", units[k].attacking)
            print("units->attack_frame: ", units[k].attack_frame)
            print("units->being_constructed: ", units[k].being_constructed)
            print("units->being_gathered: ", units[k].being_gathered)
            print("units->being_healed: ", units[k].being_healed)
            print("units->blind: ", units[k].blind)
            print("units->braking: ", units[k].braking)
            print("units->burrowed: ", units[k].burrowed)
            print("units->carrying_gas: ", units[k].carrying_gas)
            print("units->carrying_minerals: ", units[k].carrying_minerals)
            print("units->cloaked: ", units[k].cloaked)
            print("units->completed: ", units[k].completed)
            print("units->constructing: ", units[k].constructing)
            print("units->defense_matrixed: ", units[k].defense_matrixed)
            print("units->detected: ", units[k].detected)
            print("units->ensnared: ", units[k].ensnared)
            print("units->flying: ", units[k].flying)
            print("units->following: ", units[k].following)
            print("units->gathering_gas: ", units[k].gathering_gas)
            print("units->gathering_minerals: ", units[k].gathering_minerals)
            print("units->hallucination: ", units[k].hallucination)
            print("units->holding_position: ", units[k].holding_position)
            print("units->idle: ", units[k].idle)
            print("units->interruptible: ", units[k].interruptible)
            print("units->invincible: ", units[k].invincible)
            print("units->irradiated: ", units[k].irradiated)
            print("units->lifted: ", units[k].lifted)
            print("units->loaded: ", units[k].loaded)
            print("units->units->locked_down: ", units[k].locked_down)
            print("units->maelstrommed: ", units[k].maelstrommed)
            print("units->morphing: ", units[k].morphing)
            print("units->moving: ", units[k].moving)
            print("units->parasited: ", units[k].parasited)
            print("units->patrolling: ", units[k].patrolling)
            print("units->plagued: ", units[k].plagued)
            print("units->powered: ", units[k].powered)
            print("units->repairing: ", units[k].repairing)
            print("units->researching: ", units[k].researching)
            print("units->selected: ", units[k].selected)
            print("units->sieged: ", units[k].sieged)
            print("units->starting_attack: ", units[k].starting_attack)
            print("units->staised: ", units[k].stasised)
            print("units->stimmed: ", units[k].stimmed)
            print("units->stuck: ", units[k].stuck)
            print("units->targetable: ", units[k].targetable)
            print("training: ", units[k].training)
            print("units->under_attack: ", units[k].under_attack)
            print("units->under_dark_swarm: ", units[k].under_dark_swarm)
            print("units->under_disruption_web: ", units[k].under_disruption_web)
            print("units->under_storm: ", units[k].under_storm)
            print("units->upgrading: ", units[k].upgrading)

    ## RESOURCES ##
    ## It turns out that resources has up to 4 teams, but the other two are always 0, possibly filler variables
    ## In any case, we are only interested in 1v1 games
    for k, (team,resource) in enumerate(frame.resources.iteritems()):
        if k < 3:
            numResources[i,k,0] = resource.gas 
            numResources[i,k,1] = resource.ore
            numResources[i,k,2] = resource.used_psi
            numResources[i,k,3] = resource.total_psi

    ## ACTIONS ##
    if i == 0:
        print("Actions: ", frame.actions)
    
```

```python
## Visualize Map ##
theMap = rep.getMap()

## Visualize map
map1 = theMap['ground_height']
map2 = theMap['walkability']
map3 = theMap['start_locations']
map4 = theMap['buildability']

plt.imshow(map1); plt.title('ground_height'); plt.show()
plt.imshow(map2); plt.title('walkability'); plt.show()
plt.imshow(map3); plt.show()
plt.imshow(map4); plt.title('buildability'); plt.show(); 
```

```python
## Some Plots ##

## Plots the trend for number of resources in each team
plt.figure; 
plt.plot(range(numFrames), numUnits[:,0]); 
plt.plot(range(numFrames), numUnits[:,1]); plt.title('Number of Units'); plt.legend(['Team 0', 'Team 1'])
plt.show()

plt.plot(range(numFrames), numUnits[:,2]); plt.title('Number of resource units'); 
plt.show()

## Plot the trend for amount of gas and ore mined for the two teams
plt.plot(range(numFrames), numResources[:,0,0]);
plt.plot(range(numFrames), numResources[:,1,0]); plt.title('Amount of Gas Mined'); 
plt.legend(['Team 0', 'Team 1']); plt.show()

plt.plot(range(numFrames), numResources[:,0,1]);
plt.plot(range(numFrames), numResources[:,1,1]); plt.title('Amount of Ore Mined'); 
plt.legend(['Team 0', 'Team 1']); plt.show()

## Plot the trend for amount of used and total psi for the two teams
plt.plot(range(numFrames), numResources[:,0,2]);
plt.plot(range(numFrames), numResources[:,1,2]); plt.title('Used psi'); 
plt.legend(['Team 0', 'Team 1']); plt.show() 
plt.plot(range(numFrames), numResources[:,0,3]);
plt.plot(range(numFrames), numResources[:,1,3]); plt.title('Total psi'); 
plt.legend(['Team 0', 'Team 1']); plt.show()
```
